```python
import os

# --- ACADEMIC COMMAND CONFIGURATION ---
semester_dir = "Spring_Semester_2025"
subjects = ["Advanced_Organic_Chemistry", "Computational_Methods", "Molecular_Modeling"]

# Build the academic infrastructure
if not os.path.exists(semester_dir):
    os.makedirs(semester_dir)
    print(f"[+] Academic Base Constructed: ./{semester_dir}")

for subject in subjects:
    path = os.path.join(semester_dir, subject, "Syllabus_Prep")
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"   [+] Academic Sector Secured: {subject}")

# Research stays in its own specialized wing
research_dir = "Research_Acridine_Optimization"
if not os.path.exists(research_dir):
    os.makedirs(research_dir)
```
